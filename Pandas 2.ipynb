```python
#Ejercicio: encontrar las operaciones nacionales por mes de los 3 códigos IATA con
#mayor número de pasajeros internacionales

#Encuentro los 3 códigos IATA con más pasajeros internacionales
iata_inter = (pas.groupby('Codigo IATA')['Pasajeros internacionales'].sum().reset_index()
              .sort_values(by = 'Pasajeros internacionales', ascending = False).head(3))
#Ahora encuentro sus operaciones nacionales y aplico un inner join
iata_inter = (iata_inter.merge(ops.drop('Operaciones internacionales', axis=1)).
              sort_values(by = 'Anio mes'))
iata_inter.to_csv
```

```python
for col in tabla_estados.columns:
    plt.plot(tabla_estados.index, tabla_estados[col])
plt.show()
```

```python
#Ejercicio: hacer que la gráfica anterior se vea bien
tabla_estados_aux = tabla_estados.reset_index()
tabla_estados_aux['anio'] = (tabla_estados_aux['Anio mes']/100).astype(int)
tabla_estados_aux['mes'] = tabla_estados_aux['Anio mes'] - tabla_estados_aux['anio']*100
tabla_estados_aux['anio_mes'] = tabla_estados_aux['anio'] + (tabla_estados_aux['mes'] - 1)/12
tabla_estados_aux = tabla_estados_aux.drop(['anio', 'mes', 'Anio mes', 'anio_mes'], axis=1)
for col in tabla_estados_aux.columns:
    plt.plot(tabla_estados_aux.index, tabla_estados_aux[col])
plt.show()
```

```python
#Ejercicio: graficar evolución de operaciones totales de las 3 IATAs
#con más pasajeros internacionales

#Encuentro las 3 IATAs más importantes para pasajeros internacionales
#y hago join con las operaciones
tabla_iatas = (pas.groupby('Codigo IATA')['Pasajeros internacionales'].sum()
               .sort_values(ascending = False).reset_index().head(3).merge(ops.reset_index()))
#Arreglo el formato para graficar
tabla_iatas['anio'] = (tabla_iatas['Anio mes']/100).astype(int)
tabla_iatas['mes'] = tabla_iatas['Anio mes'] - tabla_iatas['anio']*100
tabla_iatas['anio_mes'] = tabla_iatas['anio'] + (tabla_iatas['mes'] - 1)/12
tabla_iatas['ops'] = tabla_iatas['Operaciones nacionales'] + tabla_iatas['Operaciones internacionales']
tabla_iatas = tabla_iatas[['anio_mes', 'ops', 'Codigo IATA']].pivot(index='anio_mes', columns='Codigo IATA', values='ops')
#Grafico
for c in tabla_iatas.columns:
    plt.plot(tabla_iatas.index, tabla_iatas[c])
plt.show()
```

```python
#Ejercicio: encontrar el promedio y desviación estándar anual de operaciones
#nacionales e internacionales de los dos estados con mayor número de pasajeros
#internacionales de 2017

ult = (pas.query('Anio_mes > 201700').groupby('Estado')['Pasajeros_internacionales'].sum().
     sort_values(ascending = False).reset_index().head(2).merge(ops).
       assign(anio = lambda x: (x['Anio_mes']/100).astype(int)).
       groupby(['Estado', 'anio'])['Operaciones_nacionales', 'Operaciones_internacionales'].
       agg([np.mean, np.std]))
ult
```
