```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
from itertools import cycle, islice
```

```python
deg = pd.DataFrame.from_csv('datos/degrees-that-pay-back.csv')
deg.head(3)
```

```python
for i in range(0, len(deg)):
    deg['Starting Median Salary'][i] = float(re.sub('[$,]', '', deg['Starting Median Salary'][i]))
```

```python
for i in range(0, len(deg)):
    deg['Mid-Career Median Salary'][i] = float(re.sub('[$,]', '', deg['Mid-Career Median Salary'][i]))
```

```python
sal_col = pd.DataFrame.from_csv('datos/salaries-by-college-type.csv')
sal_col.head(3)
```

```python
for i in range(0, len(sal_col)):
    sal_col['Starting Median Salary'][i] = float(re.sub('[$,]', '', sal_col['Starting Median Salary'][i]))
```

```python
for i in range(0, len(sal_col)):
    sal_col['Mid-Career Median Salary'][i] = float(re.sub('[$,]', '', sal_col['Mid-Career Median Salary'][i]))
```

```python
sal_reg = pd.DataFrame.from_csv('datos/salaries-by-region.csv')
sal_reg.head(3)
```

```python
for i in range(0, len(sal_reg)):
    sal_reg['Starting Median Salary'][i] = float(re.sub('[$,]', '', sal_reg['Starting Median Salary'][i]))
```

```python
for i in range(0, len(sal_reg)):
    sal_reg['Mid-Career Median Salary'][i] = float(re.sub('[$,]', '', sal_reg['Mid-Career Median Salary'][i]))
```

## Cuánto pagan diversas carreras

```python
deg[['Starting Median Salary', 'Mid-Career Median Salary']].plot(kind = 'bar', figsize = (13,3))
plt.title('Undergraduate Major Salaries', size = 20)
plt.xlabel('')
plt.show()
deg['Percent change from Starting to Mid-Career Salary'].plot(kind = 'bar', figsize = (13,3), color = 'g')
plt.title('Undergraduate Major Salary Growth', size = 20)
plt.xlabel('')
plt.show()
```

### 10 mejor pagados

```python
azules = [(0, x/10, x/10) for x in range(1,11)]
naranjas = [(x/10, x/20, 0) for x in range(1,11)]
deg['Starting Median Salary'].sort_values().tail(10).plot(kind = 'bar', figsize = (13,3), color = azules)
plt.title('Majors with Top 10 Starting Median Salary', size = 20)
plt.xlabel('')
plt.show()
deg['Mid-Career Median Salary'].sort_values().tail(10).plot(kind = 'bar', figsize = (13,3), color = naranjas)
plt.title('Majors with Top 10 Mid-Career Median Salary', size = 20)
plt.xlabel('')
plt.show()
```

### 10 peor pagados

```python
azules = [(0, x/11, x/11) for x in range(2,12)]
naranjas = [(x/11, x/21, 0) for x in range(2,12)]
deg['Starting Median Salary'].sort_values().head(10).plot(kind = 'bar', figsize = (13,3), color = azules)
plt.title('Majors with Bottom 10 Starting Median Salary', size = 20)
plt.xlabel('')
plt.show()
deg['Mid-Career Median Salary'].sort_values().head(10).plot(kind = 'bar', figsize = (13,3), color = naranjas)
plt.title('Majors with Bottom 10 Mid-Career Median Salary', size = 20)
plt.xlabel('')
plt.show()
```

### 10 con mayor crecimiento

```python
verdes = [(0, x/11, 0) for x in range(2,12)]
(deg['Percent change from Starting to Mid-Career Salary'].sort_values().tail(10)
 .plot(kind = 'bar', color = verdes, figsize = (13,5)))
plt.title('Top 10 Percentage Growth', size = 20)
plt.xlabel('')
plt.show()
```

### 10 con menor crecimiento

```python
(deg['Percent change from Starting to Mid-Career Salary'].sort_values().head(10)
 .plot(kind = 'bar', color = verdes, figsize = (13,5)))
plt.title('Bottom 10 Percentage Growth', size = 20)
plt.xlabel('')
plt.show()
```

```python
top10 = deg['Mid-Career Median Salary'].sort_values().tail(10).sum()
ot = deg['Mid-Career Median Salary'].sort_values().head(40).sum()
aux = pd.DataFrame([top10, ot], index = ['Top 10', 'Others'], columns = ['Median Salary'])
aux.plot(kind = 'pie', subplots = 'True', figsize = (10,10), autopct = '%.2f', colors =[(0.86, 0.44, 0.57), (0.85, 0.75, 0.85)])
plt.title('Top 10 Careers Salary Percentage', size = 20)
plt.ylabel('')
plt.show()
```

## Cuánto pagan según el tipo de universidad

```python
sal_col.groupby('School Type')[['Starting Median Salary', 'Mid-Career Median Salary']].apply(np.mean).sort_values(by = 'Mid-Career Median Salary').plot(kind = 'bar', figsize = (13,5))
plt.title('Salaries by School Type', size = 20)
plt.xlabel('')
plt.show()
```

```python
i = 1
plt.figure(figsize = (13,5))
for t in sal_col['School Type'].unique():
    plt.subplot(2,3,i)
    plt.hist(sal_col[sal_col['School Type'] == t]['Starting Median Salary'], bins = 5, alpha = 0.6, color = 'c', label = 'Starting')
    plt.hist(sal_col[sal_col['School Type'] == t]['Mid-Career Median Salary'], bins = 5, alpha = 0.5, color = (1, 0.6, 0), label = 'Mid-Career')
    plt.title(t, size = 15)
    plt.legend()
    i += 1
plt.tight_layout()
plt.show()
```

### 10 escuelas mejor pagadas

```python
duraznos = [(x/11, x/18, x/20) for x in range(2,12)]
(sal_col['Mid-Career Median Salary'].sort_values().tail(10)
 .plot(kind = 'pie', autopct = '%.2f', figsize = (10,10), colors = duraznos))
plt.ylabel('')
plt.title('Schools with Top 10 Median Salaries', size = 20)
plt.show()
```

```python
top10 = sal_col['Mid-Career Median Salary'].sort_values().tail(10).sum()
others = sal_col['Mid-Career Median Salary'].sort_values().head(259).sum()
aux = pd.DataFrame([top10 / (269), bot10/(269)], index = ['Top 10 Schools','Other 259'], columns = ['Median Salary'])
aux.plot(kind = 'pie', subplots = True, autopct='%.2f', figsize = (10,10), colors =[(0.86, 0.44, 0.57), (0.85, 0.75, 0.85)])
plt.ylabel('')
plt.title('Top 10 Schools Salary Percentage', size = 20)
plt.show()
```

## Cuánto pagan por región

```python
sal_reg.groupby('Region')['Mid-Career Median Salary'].apply(np.mean).plot(kind = 'bar', color = duraznos, figsize = (13,5))
plt.xlabel('')
plt.title('Mean Salary by Region', size = 20)
plt.show()
```

```python
duraznos = [(x/10, x/18, x/20) for x in range(5,11)]
(sal_reg.groupby('Region')['Mid-Career Median Salary'].apply(np.mean).sort_values()
 .plot(kind = 'pie', subplots = True, figsize = (10,10), colors = duraznos, autopct = '%.2f'))
plt.title('Salaries by Region', size = 20)
plt.ylabel('')
plt.show()
```

```python
i = 1
plt.figure(figsize = (13,5))
for t in sal_reg['Region'].unique():
    plt.subplot(2,3,i)
    plt.scatter(x = sal_reg[sal_reg['Region'] == t]['Starting Median Salary'], y = sal_reg[sal_reg['Region'] == t]['Mid-Career Median Salary'], color = duraznos, marker = '*')
    plt.title(t, size = 15)
    plt.xlabel('Starting Median Salary')
    plt.ylabel('Mid-Career Median Salary')
    i += 1
plt.tight_layout()
plt.show()
```
